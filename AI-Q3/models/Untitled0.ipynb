```python
from pytrends.request import TrendReq

# Create a TrendReq object
pytrends = TrendReq(hl='en-US', tz=360)  # Set desired language and timezone

# Build the payload for the query
kw_list = ['example keyword', 'another keyword']  # List of keywords
pytrends.build_payload(kw_list, timeframe='today 5-y')  # Set the timeframe

# Get interest over time data
interest_over_time_df = pytrends.interest_over_time()

# Print the retrieved data
print(interest_over_time_df)
```

```python
from pytrends.request import TrendReq

# Create a TrendReq object
pytrends = TrendReq(hl='en-US', tz=360)  # Set desired language and timezone

# Take input for keywords
kw_list = input("Enter keywords (comma-separated): ").split(',')

# Build the payload for the query with the specified keywords
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y')

# Get interest over time data
interest_over_time_df = pytrends.interest_over_time()

# Print the retrieved data
print(interest_over_time_df)
```

```python
import requests
from bs4 import BeautifulSoup

# Function to get real-time trends for a specific country
def get_realtime_trends(country_code):
    # Construct the URL for Google Trends real-time data
    url = f"https://trends.google.com/trends/trendingsearches/realtime?geo={country_code}&hl=en-US&category=all"

    # Send a GET request to the URL
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find and extract the trending searches
    trending_searches = []
    search_items = soup.find_all("div", class_="title")
    for item in search_items:
        trending_searches.append(item.get_text())

    return trending_searches

# Provide the desired country code
country_code = "IN"  # Change this to the desired country code

# Get real-time trends using the function
trends = get_realtime_trends(country_code)

# Print the list of trends
for i, trend in enumerate(trends, start=1):
    print(f"{i}. {trend}")
```

```python
import gradio as gr
import requests
from bs4 import BeautifulSoup
from pytrends.request import TrendReq

# Function to get real-time trends for a specific country
def get_realtime_trends(country_code):
    # Construct the URL for Google Trends real-time data
    url = f"https://trends.google.com/trends/trendingsearches/realtime?geo={country_code}&hl=en-US&category=all"

    # Send a GET request to the URL
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find and extract the trending searches
    trending_searches = []
    search_items = soup.find_all("div", class_="title")
    for item in search_items:
        trending_searches.append(item.get_text())

    return trending_searches

# Gradio interface function
def gr_interface(country_code="IN"):
    # Get real-time trends using the function
    trends = get_realtime_trends(country_code)

    # Combine the trends into a single string
    trends_string = "\n".join(trends)

    return trends_string

iface = gr.Interface(fn=gr_interface, inputs="text", outputs="text", title="Google Trends Real-Time Search", description="Enter a country code to get real-time trending searches.")
iface.launch()
```
